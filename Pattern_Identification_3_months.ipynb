```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('COST (1).csv')

# Convert the Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Filter the DataFrame to include only rows with dates after 05/01/2024
filtered_df = df[df['Date'] >= '2024-05-01']

# Display the filtered DataFrame

filtered_df
```

```python
df= filtered_df[['Date', 'Open', 'High', 'Low', 'Close']]
df
```

```python
df['Body'] = abs(df['Close'] - df['Open'])
df['CandleSize'] = df['High'] - df['Low']

# Identify bullish green candlestick patterns
df['BullishGreen'] = (df['Close'] > df['Open']) & (df['Body'] >= 0.5 * df['CandleSize'])

# Display the DataFrame with the new column
df
```

```python
df['Body'] = abs(df['Close'] - df['Open'])
df['CandleSize'] = df['High'] - df['Low']

# Identify bullish green candlestick patterns
df['BearRead'] = (df['Close'] < df['Open']) & (df['Body'] >= 0.5 * df['CandleSize'])

# Display the DataFrame with the new column
df
```

```python
max_body_size = 0.0004
min_candle_size = 0.01

# Load the stock price data from CSV
# df = pd.read_csv('COST (1).csv')

# Calculate the necessary conditions
df['BodySize'] = abs(df['Open'] - df['Close'])
df['CandleSize'] = df['High'] - df['Low']

# Identify Doji pattern
df['Doji'] = ((df['BodySize'] < df['Close'] * max_body_size) &
              (df['CandleSize'] > df['Close'] * min_candle_size))

# Filter only the Doji rows
df
```

```python
df['UpperShadow'] = df['High'] - df[['Open', 'Close']].max(axis=1)
df['LowerShadow'] = df[['Open', 'Close']].min(axis=1) - df['Low']

# Identify spinning top candlestick patterns
df['SpinningTop'] = ((~df['Doji']) & df['Body'] <= 0.20 * df['CandleSize']) & \
                    (df['UpperShadow'] >= 0.25 * df['CandleSize']) & \
                    (df['LowerShadow'] >= 0.25* df['CandleSize'])

# Display the DataFrame with the new column
df
```

```python
df['Body'] = abs(df['Close'] - df['Open'])
df['CandleSize'] = df['High'] - df['Low']

# Calculate the lower and upper shadows
df['LowerTail'] = df[['Open', 'Close']].min(axis=1) - df['Low']
df['UpperTail'] = df['High'] - df[['Open', 'Close']].max(axis=1)

# Identify Hammer candlestick patterns
df['Hammer'] = (
    (df['LowerTail'] >= 2 * df['Body']) &        # Lower tail is at least twice the body
    (df['UpperTail'] <= 0.3 * df['CandleSize']) &  # Upper tail is relatively small
    (df['Body'] <= 0.3 * df['CandleSize'])       # Body is relatively small
)

# Display the DataFrame with the new column
df
```

```python
df['Body'] = abs(df['Close'] - df['Open'])
df['CandleSize'] = df['High'] - df['Low']

# Calculate the lower and upper shadows
df['LowerTail'] = df[['Open', 'Close']].min(axis=1) - df['Low']
df['UpperTail'] = df['High'] - df[['Open', 'Close']].max(axis=1)

# Identify Inverted Hammer candlestick patterns
df['InvertedHammer'] = (
    (df['LowerTail'] <= 0.3 * df['CandleSize']) &  # Lower tail is very small
    (df['UpperTail'] >= 2 * df['Body']) &        # Upper tail is at least twice the body
    (df['Body'] <= 0.3 * df['CandleSize'])       # Body is relatively small
)

# Display the DataFrame with the new column
print(df)
```
